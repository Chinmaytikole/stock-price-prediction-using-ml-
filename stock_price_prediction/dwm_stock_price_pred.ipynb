```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import yfinance as yf
import pandas as pd
import pandas_ta as ta
```

```python
stock = "HAVELLS.NS"
VBL_data = yf.download(stock, start="2022-09-11",end='2024-09-11')
VBL_data
```

```python
vbl_close = pd.DataFrame()
VBL_data['Close'] = VBL_data['Adj Close']
VBL_data['change'] = VBL_data['Close'].pct_change()
VBL_data.dropna()
```

```python
VBL_data['Close'].plot(figsize = (15,8))
plt.show()
```

```python
VBL_data['LEMA'] = ta.ema(VBL_data['Close'],length=10)
VBL_data['RSI'] = ta.rsi(VBL_data['Close'],length = 14)
VBL_data
```

```python

#plt.style.use("seaborn")
VBL_data['Close'].div(VBL_data['Close'][0]).mul(100).plot(figsize = (15,8))
VBL_data['LEMA'].div(VBL_data['LEMA'][10]).mul(100).plot(figsize = (15,8))
VBL_data['RSI'].div(VBL_data['RSI'][14]).mul(100).plot(figsize = (15,8))
plt.show()
```

```python
VBL_data = VBL_data.dropna()
```

```python
import numpy as np
VBL_data['ema_class'] = np.where(VBL_data['LEMA'] > VBL_data['Close'], 1, 0)
```

```python
m = (VBL_data["RSI"].max() + VBL_data["RSI"].min())/2
VBL_data['rsi_class'] = np.where(VBL_data["RSI"] >m, 1, 0)
pd.set_option('display.max_rows', None)
```

```python
VBL_data['next_cls'] = VBL_data["Close"].shift(-1)
```

```python
VBL_data['cls_class'] = np.where(VBL_data["Close"] < VBL_data["next_cls"] , 1, 0)
pd.set_option('display.max_rows', None)
```

```python
vbl_class = VBL_data[["Low","High",'Close','ema_class','rsi_class','cls_class']]
```

```python
supertrend_output = ta.supertrend(VBL_data["High"], VBL_data["Low"], VBL_data["Close"], length=10, multiplier=3)
supertrend_output
```

```python

# Check what the output looks like
# print(supertrend_output)
vbl_class["direct"] = supertrend_output['SUPERTd_10_3.0']
vbl_class['super'] = supertrend_output['SUPERT_10_3.0']

vbl_class = vbl_class.dropna()
VBL_data.drop(index='2022-10-03').reset_index(drop=True)
vbl_class
```

```python
# Step 1: Import necessary libraries
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn import metrics
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load your dataset
# Assuming vbl_class is already defined and contains your data

# Prepare features and target
X = vbl_class[['Low', 'High', 'Close']].dropna()
y = vbl_class['cls_class']  # Target

# Step 3: Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Step 4: Create an ID3 Decision Tree Classifier
id3_model = DecisionTreeClassifier(criterion='entropy')

# Step 5: Train the model
id3_model.fit(X_train, y_train)

# Step 6: Make predictions
y_pred = id3_model.predict(X_test)

# Step 7: Evaluate the model
accuracy = metrics.accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

y_pred = id3_model.predict([X_test])

# Step 8: Visualize the Decision Tree
plt.figure(figsize=(20, 10))
# Convert class names to strings
class_names = [str(cls) for cls in sorted(y.unique())]
plot_tree(id3_model, filled=True, feature_names=X.columns, class_names=class_names, rounded=True)
plt.title("Decision Tree Visualization")
plt.show()
```
